### 전처리
- 전통적인 머신러닝 알고리즘 성능에 데이터 특성에 따른 **전처리**가 중요

  - normalization
  - standardization
- overfitting 걸렸을때 feature selection, dim reduction(차원축소는 대표성을 유지함)
  > ##### feature selection
    1. filter method
    2. wrapper mthod
    3. embeded 
- 3차원 데이터 1차원으로 만들 수 있음
- LDA, PCA 성능을 유지한채 과적합 줄임
  - > **차원의 저주(curse of dimensionality)** : 데이터 양에 비해 차원(column)이 많음
  - overfitting : 학습한 데이터에서만 성능이 좋음
