### Numpy 함수형 패러다임

- [Functional Programming](https://numpy.org/doc/stable/reference/routines.functional.html)
   - 기존의 함수 재사용해서 새로운 함수를 만듦

```
oct_array = np.frompyfunc(oct, 1, 1)
oct_array(np.array((10, 30, 100)))  #함수형 패러다임 관점
```

```
t=np.vectorize(lambda x,y:x+y)#pyfunc 인자로 받음
t([2],[3]) #원소의 합
```

#### @np.vectorize
데코레이터 방식도 지원

```
@np.vectorize
def tt(x,y):
  return x+y
```

```
tt([2],[3])
```
